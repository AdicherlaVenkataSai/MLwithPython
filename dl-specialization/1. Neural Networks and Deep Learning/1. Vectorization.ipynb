# Demo

```
import time
import numpy as np

a = np.random.rand(1000000)
b = np.random.rand(1000000)

t1 = time.time()
c = np.dot(a, b)
t2 = time.time()
print('Vectorized version:' + str(1000 *(t2 - t1)) + 'ms')
# .dot(which dont have for loops) enables parallel computation so GPU or TPU make code run faster
# note: whenever  possible avoid explicit for loops
t1 = time.time()
c = 0
for i in range(1000000):
  c += a[i] * b[i]
t2 = time.time()
print('Non-vectorized version:' + str(1000 *(t2 - t1)) + 'ms')
```
