# Tips and Tricks
## 1. Don't use rank one arrays

-  don't ( np.random.randn(5) )
-  do (np.random.randn(5, 1) or (1, 5) )

note: 
1. use asserstation statement to check the vectors
2. if we encounter the rank 1 arrays, we can reshape it a.reshape((5, 1) or (1, 5))


```
import numpy as np

a = np.random.randn(5)
# assert(a.shape == (5, 1))
print(a)
# rank 1 array
```

```
print(a.shape)
```

note: This neither a row vector or a col vector, but its a rank 1 array

```
print(a.T)
# see we have same values for both a and a.T
```

```
print(np.dot(a, a.T))
# we hope we'll a  matrix rather it end up with a scalar value
```

```
a = np.random.randn(5,1)
a
# a is col vector
```

```
a.T
# a.T would be a row vector
```

```
print(np.dot(a, a.T))
# now we'll get a matrix
```
