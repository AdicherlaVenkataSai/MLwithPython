-  we fit the line y = mx + c, but which line fits perfectly
-  so we calculate the residual errors for every line and choose the one with min error
-  error = sum(observed - predicted)^2
 
Fitting the data with line is it a good choice
-  what if the data is not linear? if not then quadratic
-  why not quadratic function? if so find the best quadratic with min residual error  
-  y = c + x1m1 + x2m2 (This is still called linear regression with an additional dimension)
 
Can we try other higher order polynomials ?
-  May be yes/ no, mostly we don’t choose it because of overfitting the data
-  if we do, can it be a generalised fit? it might result in bad predictions
 
How do we choose model order/ complexity?
-  simulate the predictions (
 
  a. remove some houses 
 
  b. fit model on remaining
 
  c. predict held out houses)
-  then compare them with actual values, then we get to know about its predictions
 
Note: To generalize it we can say train, test sets (Training error, Test error)
 
Training error: decreases as the order goes on increase
 
Test error: decreases to some order and keeps increasing after certain order, as the model complexity increases

Note 3: in the module we discussed residual sum of squares (RSS) as an error metric for regression, but Turi Create uses root mean squared error (RMSE). These are two common measures of error regression, and RMSE is simply the square root of the mean RSS:


where N is the number of data points. RMSE can be more intuitive than RSS, since its units are the same as that of the target column in the data, in our case the unit is dollars ($), and doesn't grow with the number of data points, like the RSS does.
